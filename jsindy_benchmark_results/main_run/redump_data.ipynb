```python
import pickle
import jax.numpy as jnp
import jax
jax.config.update('jax_enable_x64',True)
jax.config.update('jax_default_device',jax.devices()[0])

import pathlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Code I used to put all the pickles together

```python
f1_vals = {}
l2_error = {}

file_paths = list(pathlib.Path('results').glob("*.pkl"))
def get_properties(path):
	pieces = str(path).split('_')
	run = pieces[2]
	seed_val = pieces[0][-1]
	noise_level = pieces[5]
	time_length = pieces[-1][:-4]
	return run,seed_val,float(noise_level),float(time_length)


all_results = dict()
for pkl_file in file_paths:
	properties = get_properties(pkl_file)
	with open(pkl_file, 'rb') as file:
		results = pickle.load(file)
	all_results[properties] = results

with open('all_results.pkl', 'wb') as file:
    pickle.dump(all_results, file)
```

```python
with open('all_results.pkl','rb') as f:
    all_results = pickle.load(f)
```

```python
f1_vals = {key:result['coeff_mets']['f1'] for key,result in all_results.items()}
l2_error = {key:result['coeff_mets']['coeff_rel_l2'] for key,result in all_results.items()}
main_results = {'f1':f1_vals,'l2':l2_error}
with open('main_results.pkl','wb') as f:
    pickle.dump(main_results,f)
```
